```python
import matplotlib.pyplot as plt
import geopandas as gpd
import os

plt.style.use("ggplot")
```

```python
out_folder = r"../data/spatial"

gdf = gpd.read_file(r"../data/spatial/Sidenedborfelt.shp")
# vassom_gdf = vassom_gdf.to_crs('epsg:25833')
gdf
```

```python
nedborfelt_li = ['Barduelva', 'Divielva']

gdf_dict = {}

for riv in nedborfelt_li:
    cur_gdf = gdf.loc[gdf['nedborfelt'] == riv]
    out_fpath = os.path.join(out_folder, "Sidenedborfelt_%s.shp" % riv)
    cur_gdf.to_file(out_fpath)
    gdf_dict[riv] = cur_gdf
```
