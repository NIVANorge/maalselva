```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fpath = r'../data/remote_sensing/Maalselv_wetsnow.txt'
df = pd.read_csv(fpath, sep='\t',
                 index_col=0, parse_dates=True)
df = df.round(3)
df
```

```python
df['Sum'] = df.sum(axis=1)
df
```

```python
df.plot(subplots=True, figsize=(15, 12), grid=True, sharex=True, sharey=True)
plt.savefig(r'../figures/snowcover/snowcover_ts.png')
```
